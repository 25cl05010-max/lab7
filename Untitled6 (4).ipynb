### lab 7 Calculate the sample mean and plot the histgram

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
#path to the folder 
path="C:\\Users\\user\\Desktop\\25CL05010\\17_oct_cda\\HadISST_sst.nc"
ds=xr.open_dataset(path)
ds
```

```python
#Defining the variable
sst=ds['sst']
```

```python
#extracting the EL NINO3.4 area
lat_long=sst.sel(longitude=slice(-170,-120) , latitude=slice(5,-5))
lat_long
```

```python
#domain average
steps=lat_long.mean(dim=['latitude','longitude'])
steps
```

# calculating mean of random 360 steps, drawing 1000 times

```python
#using for loop function 
data=np.zeros([1000,1])
for i in range(1,1000):
    xx = np.random.choice(steps,360, replace='false')
    data[i] = np.mean(xx)
    
    
print(data)
```

# population mean

```python
#calculating mean
mean=np.mean(steps)
print(mean)
```

# plotting histogram for 1000 mean values

```python
#plotting histogram
plt.figure(figsize=(9,4))
#add the vertical line
Mean= 26.93798
plt.axvline(Mean, color='red', linestyle='--', linewidth=2, label='population mean')
plt.hist(data[1:], bins='auto', alpha=0.6, label='Sample mean', edgecolor='black')

plt.xlabel("Mean values")
plt.ylabel("Frequency")
plt.title("Distribution of sample mean")
plt.legend()
plt.tight_layout()
plt.show()





```

# mean follows normal distribution, so it gives bell shaped curve.
